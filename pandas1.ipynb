## 使用pivot

```python
df.head()
```

```python
df.shape
```

```python
df2 = df.reset_index()
df2 = df2[:16].copy()
df2.shape
```

```python
df2['n'] = df2.index
df2
```

```python
df2['cat1'] = df2['n'].apply(lambda x: x%2)
df2['cat2'] = df2['n'].apply(lambda x: x%4//2)
df2['cat3'] = df2['n'].apply(lambda x: x%8//4)
df2['cat4'] = df2['n'].apply(lambda x: x%16//8)
```

```python
df2
```

```python
df2.pivot(['cat1','cat3'],['cat2','cat4'])
```

```python
df2[:4].pivot('cat1','cat2','Country')
```

```python
df2[:8].pivot(['cat1','cat2'],'cat3','Country')
```
